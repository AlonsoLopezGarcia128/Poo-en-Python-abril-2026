#### Conceptos POO (Programacion orientada a objetos) en Python

### El prompt en Gemini IA
## Lenguaje Python, crear la clase "animal" con los atributos "nombre", "raza"; crear el constructor animal, y las funciones ladrar(), correr(), crear los objetos pitbull, chihuahua, nivel principiante, no es interactivo, proporcionar tu los datos

```python
class Animal:
    # El constructor __init__ inicializa los atributos de cada objeto
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función para que el animal ladre
    def ladrar(self):
        print(f"{self.nombre} dice: ¡Guau, guau!")

    # Función para que el animal corra
    def correr(self):
        print(f"{self.nombre} está corriendo velozmente.")

    def morder(self):
        print(f"{self.nombre} está mordiendo a la vecina")

# --- Creación de Objetos (Instancias) ---

# Creamos el objeto 'pitbull' con datos específicos
pitbull = Animal("Rocco", "Pitbull")

# Creamos el objeto 'chihuahua' con datos específicos
chihuahua = Animal("Chispa", "Chihuahua")
# Creamos el objeto perro callejero con datos especificos
Chafa= Animal("solovino", "Chafa")


# --- Uso de los objetos ---

print(f"Perro 1: {pitbull.nombre}, Raza: {pitbull.raza}")
pitbull.ladrar()
pitbull.correr()
pitbull.morder()

print("-" * 20)

print(f"Perro 2: {chihuahua.nombre}, Raza: {chihuahua.raza}")
chihuahua.ladrar()
chihuahua.correr()
chihuahua.morder()

print("-" * 20)

print(f"Perro 3: {Chafa.nombre}, Raza: {Chafa.raza}")
Chafa.ladrar()
Chafa.correr()
Chafa.morder()

print("Programa creado por Alonso López García Nl = 31")
```
